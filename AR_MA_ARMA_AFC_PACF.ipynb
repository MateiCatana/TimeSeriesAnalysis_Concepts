```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Generm o serie de date de exemplu (poate fi nlocuit cu propriile dvs. date)
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=100)

# Plotm funcia de autocorelaie (ACF)
plot_acf(data, lags=20)
plt.title('Funcia de autocorelaie (ACF)')
plt.xlabel('Lag')
plt.ylabel('Autocorelaie')
plt.show()

# Plotm funcia de autocorelaie parial (PACF)
plot_pacf(data, lags=20)
plt.title('Funcia de autocorelaie parial (PACF)')
plt.xlabel('Lag')
plt.ylabel('Autocorelaie parial')
plt.show()
```
