```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Generăm o serie de date de exemplu (poate fi înlocuită cu propriile dvs. date)
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=100)

# Plotăm funcția de autocorelație (ACF)
plot_acf(data, lags=20)
plt.title('Funcția de autocorelație (ACF)')
plt.xlabel('Lag')
plt.ylabel('Autocorelație')
plt.show()

# Plotăm funcția de autocorelație parțială (PACF)
plot_pacf(data, lags=20)
plt.title('Funcția de autocorelație parțială (PACF)')
plt.xlabel('Lag')
plt.ylabel('Autocorelație parțială')
plt.show()
```
