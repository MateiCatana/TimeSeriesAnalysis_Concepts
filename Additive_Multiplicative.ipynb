```python
import numpy as np
import matplotlib.pyplot as plt

# Crearea datelor de timp
time = np.arange(1, 25)  # 12 luni

# Definirea trendului (cretere liniar)
trend = 2 * time

# Definirea componentei sezoniere (constant)
seasonality_additive = np.ones_like(time) * 5

# Definirea componentei sezoniere (variabil)
seasonality_multiplicative = np.sin(time / 2 * np.pi) * 5 + 5  # Variaie sinusoidal cu amplitudine 5 i medie 5

# Crearea seriei de timp aditive
additive_series = trend + seasonality_additive

# Crearea seriei de timp multiplicative
multiplicative_series = trend * seasonality_multiplicative

# Crearea graficului pentru modelul aditiv
plt.figure(figsize=(10, 5))
plt.plot(time, additive_series, label='Aditiv', marker='o')
plt.plot(time, trend, label='Trend', linestyle='--')
plt.plot(time, seasonality_additive, label='Sezon', linestyle='--')
plt.xlabel('Timp')
plt.ylabel('Valoare')
plt.title('Model Aditiv')
plt.legend()
plt.grid(True)
plt.show()

# Crearea graficului pentru modelul multiplicative
plt.figure(figsize=(10, 5))
plt.plot(time, multiplicative_series, label='Multiplicativ', marker='o')
plt.plot(time, trend, label='Trend', linestyle='--')
plt.plot(time, seasonality_multiplicative, label='Sezon', linestyle='--')
plt.xlabel('Timp')
plt.ylabel('Valoare')
plt.title('Model Multiplicativ')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generm date pentru seria de timp determinist (micarea unui pendul)
t = np.linspace(0, 10, 100)  # timpul de la 0 la 10 secunde
L = 1  # lungimea firului pendulului
g = 9.81  # acceleraia gravitaional
theta0 = np.pi / 4  # unghiul iniial de deviere (45 de grade)
theta = theta0 * np.cos(np.sqrt(g / L) * t)  # ecuaia micrii pendulului

# Generm date pentru seria de timp stohastic (evoluia preurilor ntr-o pia financiar)
np.random.seed(0)  # stabilim o smn pentru reproducibilitate
prices = 100 + np.cumsum(np.random.randn(100))  # preurile sunt influenate de zgomot aleatoriu

# Generm graficul pentru seria de timp determinist (micarea unui pendul)
plt.figure(figsize=(10, 5))
plt.plot(t, theta, label='Deterministic', color='blue')
plt.xlabel('Timp')
plt.ylabel('Unghi (radiani)')
plt.title('Serie de timp determinist: Micarea unui pendul')
plt.legend()
plt.grid(True)
plt.show()

# Generm graficul pentru seria de timp stohastic (evoluia preurilor ntr-o pia financiar)
plt.figure(figsize=(10, 5))
plt.plot(prices, label='Stochastic', color='red')
plt.xlabel('Timp')
plt.ylabel('Preuri')
plt.title('Serie de timp stohastic: Evoluia preurilor ntr-o pia financiar')
plt.legend()
plt.grid(True)
plt.show()
```
