```python
import matplotlib.pyplot as plt

# Datele seriei temporale (valori fictive pentru exemplu)
timp = range(1, 11)
valori = [10, 12, 14, 16, 18, 20, 23, 25, 27, 29]

# Desenarea graficului
plt.plot(timp, valori, marker='o', linestyle='-')

# Adăugarea unui punct de ruptură în trend (schimbarea bruscă a direcției)
plt.plot(6, 20, marker='*', markersize=10, color='red')

# Adăugarea titlului și etichetelor pentru axe
plt.title('Punct de ruptură în trend')
plt.xlabel('Timp')
plt.ylabel('Valoare')

# Afișarea graficului
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datele seriei temporale (valori fictive pentru exemplu)
timp = range(1, 16)
valori = [10, 12, 14, 16, 18, 20, 10, 12, 14, 10, 12, 14, 16, 18, 20]

# Desenarea graficului
plt.plot(timp, valori, marker='o', linestyle='-')

# Adăugarea a trei puncte de ruptură în trend (schimbarea bruscă a direcției)
plt.plot(6, 10, marker='*', markersize=10, color='red')
plt.plot(10, 20, marker='*', markersize=10, color='red')
plt.plot(14, 10, marker='*', markersize=10, color='red')

# Adăugarea titlului și etichetelor pentru axe
plt.title('Puncte de ruptură în trend')
plt.xlabel('Timp')
plt.ylabel('Valoare')

# Afișarea graficului
plt.grid(True)
plt.show()
```
