```python
import numpy as np
import matplotlib.pyplot as plt

# Generm date pentru media, variana i autocovariana
t = np.arange(0, 100)  # Timpul
mu_t = 2 * t  # Funcia de medie: t = 2t
sigma_t_squared = 3 * t  # Funcia de varian: ^2_t = 3t
gamma_k = 1 / (1 + np.abs(t))  # Funcia de autocovarian: (k) = 1 / (1 + |k|)

# Crem graficul
plt.figure(figsize=(10, 6))

# Medie
plt.subplot(3, 1, 1)
plt.plot(t, mu_t, color='blue')
plt.title('Comportamentul mediei, varianei i autocovarianei n funcie de timp')
plt.ylabel('Medie')
plt.grid(True)

# Varian
plt.subplot(3, 1, 2)
plt.plot(t, sigma_t_squared, color='green')
plt.ylabel('Varian')
plt.grid(True)

# Autocovarian
plt.subplot(3, 1, 3)
plt.plot(t, gamma_k, color='red')
plt.xlabel('Timp')
plt.ylabel('Autocovarian')
plt.grid(True)

plt.tight_layout()
plt.show()
```
