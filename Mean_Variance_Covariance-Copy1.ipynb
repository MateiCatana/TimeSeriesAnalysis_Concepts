```python
import numpy as np
import matplotlib.pyplot as plt

# Generăm date pentru media, varianța și autocovarianța
t = np.arange(0, 100)  # Timpul
mu_t = 2 * t  # Funcția de medie: μt = 2t
sigma_t_squared = 3 * t  # Funcția de varianță: σ^2_t = 3t
gamma_k = 1 / (1 + np.abs(t))  # Funcția de autocovarianță: γ(k) = 1 / (1 + |k|)

# Creăm graficul
plt.figure(figsize=(10, 6))

# Medie
plt.subplot(3, 1, 1)
plt.plot(t, mu_t, color='blue')
plt.title('Comportamentul mediei, varianței și autocovarianței în funcție de timp')
plt.ylabel('Medie')
plt.grid(True)

# Varianță
plt.subplot(3, 1, 2)
plt.plot(t, sigma_t_squared, color='green')
plt.ylabel('Varianță')
plt.grid(True)

# Autocovarianță
plt.subplot(3, 1, 3)
plt.plot(t, gamma_k, color='red')
plt.xlabel('Timp')
plt.ylabel('Autocovarianță')
plt.grid(True)

plt.tight_layout()
plt.show()
```
