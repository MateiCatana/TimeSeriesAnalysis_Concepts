```python
import numpy as np
import matplotlib.pyplot as plt

# Setăm sămânța pentru reproducibilitate
np.random.seed(42)

# Generăm date pentru seria de timp cu stationaritate strictă
stationary_data = np.random.normal(loc=0, scale=1, size=1000)

# Generăm date pentru seria de timp cu stationaritate în sens larg
non_stationary_data = np.cumsum(np.random.normal(loc=0, scale=1, size=1000))

# Generăm date pentru zgomotul alb (white noise)
white_noise_data = np.random.normal(loc=0, scale=1, size=1000)

# Creăm graficul pentru seria de timp cu stationaritate strictă
plt.figure(figsize=(10, 5))
plt.plot(stationary_data, color='blue')
plt.title('Serie de timp cu stationaritate strictă')
plt.xlabel('Timp')
plt.ylabel('Valoare')
plt.grid(True)
plt.show()

# Creăm graficul pentru seria de timp cu stationaritate în sens larg
plt.figure(figsize=(10, 5))
plt.plot(non_stationary_data, color='red')
plt.title('Serie de timp cu stationaritate în sens larg')
plt.xlabel('Timp')
plt.ylabel('Valoare')
plt.grid(True)
plt.show()

# Creăm graficul pentru zgomotul alb (white noise)
plt.figure(figsize=(10, 5))
plt.plot(white_noise_data, color='green')
plt.title('Zgomot alb (White Noise)')
plt.xlabel('Timp')
plt.ylabel('Valoare')
plt.grid(True)
plt.show()
```
