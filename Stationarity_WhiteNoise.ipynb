```python
import numpy as np
import matplotlib.pyplot as plt

# Setm smna pentru reproducibilitate
np.random.seed(42)

# Generm date pentru seria de timp cu stationaritate strict
stationary_data = np.random.normal(loc=0, scale=1, size=1000)

# Generm date pentru seria de timp cu stationaritate n sens larg
non_stationary_data = np.cumsum(np.random.normal(loc=0, scale=1, size=1000))

# Generm date pentru zgomotul alb (white noise)
white_noise_data = np.random.normal(loc=0, scale=1, size=1000)

# Crem graficul pentru seria de timp cu stationaritate strict
plt.figure(figsize=(10, 5))
plt.plot(stationary_data, color='blue')
plt.title('Serie de timp cu stationaritate strict')
plt.xlabel('Timp')
plt.ylabel('Valoare')
plt.grid(True)
plt.show()

# Crem graficul pentru seria de timp cu stationaritate n sens larg
plt.figure(figsize=(10, 5))
plt.plot(non_stationary_data, color='red')
plt.title('Serie de timp cu stationaritate n sens larg')
plt.xlabel('Timp')
plt.ylabel('Valoare')
plt.grid(True)
plt.show()

# Crem graficul pentru zgomotul alb (white noise)
plt.figure(figsize=(10, 5))
plt.plot(white_noise_data, color='green')
plt.title('Zgomot alb (White Noise)')
plt.xlabel('Timp')
plt.ylabel('Valoare')
plt.grid(True)
plt.show()
```
